```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_excel('1.4_02.xlsx')
```

```python
df.columns = ['region1', 'agriculture', 'industry', 'construction', 'services', 'government']
df.head()
```

```python
df["region"] = df["region1"].str.split(',').apply(lambda x: x[0])
df["year"] = df["region1"].str.split(',').apply(lambda x: x[1])
```

```python
df.head()
```

```python
new_df = []
for ind, row in df.iterrows():
    for c in ["agriculture", "industry", "construction", "services", "government"]:
        r = {}
        r["region"] = row["region"]
        r["year"] = int(row["year"])
        r["department"] = c
        r["percent"] = row[c]
        new_df.append(r)

new_df = pd.DataFrame(new_df)
```

```python
new_df.head()
```

```python
ukraine = new_df[(new_df['region'] == 'Україна')]
ukraine
```

```python
alt.Chart(new_df).mark_bar().encode(
    x = alt.X('percent:Q', aggregate = 'sum', sort=alt.Sort(field="percent")),
    y = alt.Y('year:N'),
    order=alt.Order(
      # Sort the segments of the bars by this field
      'percent',
      sort='descending'
    ),
    color = alt.Color('department:N'),
    row = alt.Row('region:N'),
    tooltip = [alt.Tooltip('department'), alt.Tooltip('percent')],
).properties(width = 950, height = 500)
```
