```python
import pandas as pd
import altair as alt
```

Проблема візуалізації з екселю в тому, що вона не дозволяє побачити як саме змінилася структура ВДВ - дуже складно порівнювати стовпчики за висотою коли вони не стоять поруч - тому складно сказати яка саме галузь переважає в конкретній області. Також неможливо порівняти чи виросла чи ні конкретна галузь з 2010 по 2017 рік - бо неможливо порівняти відсотки конкретної галузі, адже їх не видно. 

```python
df = pd.read_excel('1.4_02.xlsx')
```

```python
df.columns = ['region1', 'agriculture', 'industry', 'construction', 'services', 'government']
```

```python
df["region"] = df["region1"].str.split(',').apply(lambda x: x[0])
df["year"] = df["region1"].str.split(',').apply(lambda x: x[1])
```

```python
new_df = []
for ind, row in df.iterrows():
    for c in ["agriculture", "industry", "construction", "services", "government"]:
        r = {}
        r["region"] = row["region"]
        r["year"] = int(row["year"])
        r["type"] = c
        r["percent"] = row[c]
        new_df.append(r)

new_df = pd.DataFrame(new_df)
```

**На даній візуалізації видно як саме змінилася структура ВДВ у кожній області - адже всі галузі посортовані по їх відсотковій кількості. Для того щоб дізнатися відсоток конкретної галузі я додала тултіп до кожного прямокутника - з назвою галузі та його відсотком. Хоч на цій візуалізації складно візуально оцінити як змінився відсоток конкретної галузі, але вона дозволяє оцінити саме структуру ВДВ в області загалом.**

```python
alt.Chart(new_df).mark_bar().encode(
    x = alt.X('percent:Q', aggregate = 'sum', sort=alt.Sort(field="percent")),
    y = alt.Y('year:N'),
    order=alt.Order(
      # Sort the segments of the bars by this field
      'percent',
      sort='descending'
    ),
    color = alt.Color('type:N'),
    row = alt.Row('region:N'),
    tooltip = [alt.Tooltip('type'), alt.Tooltip('percent')],
).properties(width = 950, height = 100)
```

**Для того щоб оцінити як за 7 років змінилася кожна окрема галузь я зробила візуалізацію, на якій просто порівнюється відсоток кожної галузі. Але її проблема в тому, що вона не дозволяє оцінити саме структуру ВДВ - тобто ми не бачимо суму галузей в цілому. Тому ця і попередня візуалізація доповнюють одна іншу.**

```python
alt.Chart(new_df).mark_bar().encode(
    x = alt.X('percent'),
    y = alt.Y('year:N'),
    row = alt.Row('region:N'),
    column = alt.Row('type:N'),
    tooltip = [alt.Tooltip('type'), alt.Tooltip('percent')],
    color = alt.Color('type:N')
).properties(width = 125, height = 50)
```
